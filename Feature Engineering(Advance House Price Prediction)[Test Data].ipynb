## Missing values

```python
## Let us capture all the nan values
## First Lets handle categorical features which are missing
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
for feature in features_nan:
    print(f"{feature}:{np.round(dataset[feature].isnull().mean(),4)} % missing values")
```

```python
## Replace Missing values with a new label
def replace_cat_feature(dataset,features_nan):
    data = dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
```

```python
dataset = replace_cat_feature(dataset,features_nan)
```

```python
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
## Now lets check for numerical features that contains missing values
numerical_features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']

## We will print the numerical nan features and percentage of missing values
for feature in numerical_features_nan:
    print(f"{feature}:{np.round(dataset[feature].isnull().mean(),4)}% missing values")
```

```python
## replacing the numerical missing features
for feature in numerical_features_nan:
    ## we will replace by using median since there are outliers
    median_value = dataset[feature].median()
    
    ## create new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```

## Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
dataset.head()
```

```python
import numpy as np
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
    
```

## Handling Rare Categorical Feature

We will reomove categorical variables that are present less than 1% of the observations

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
categorical_feature
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SaleCondition'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(1)
```

```python
for feature in categorical_feature:
    labels_ordered = dataset.groupby([feature])['SaleCondition'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
    
```
