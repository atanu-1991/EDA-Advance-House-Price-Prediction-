## Missing Values

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

```python
for feature in features_with_na:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### Numerical Variables

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print('Number of Numerical Features: ',len(numerical_features))
dataset[numerical_features].head()
```

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
print(year_feature)
```

```python
# Let's explore the content of these year variables
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
## Letz analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
year_feature
```

```python
## Here we will compare the difference between All the year feature with Sale Price
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        ## we will capture the difference between year the house was sold and year variable
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature+' vs SalePrice')
        plt.show()
```

```python
len(dataset['SalePrice'].unique())
```

```python
## Numerical variables are usually of 2 types
## 1. Continous variable and 2. Discrete variable

discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print(f"Discrete feature Count: {len(discrete_feature)}")
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Lets find the relationship between discrete_feature and sale price
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

#### Continuous Feature

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print('Continous Feature Count: ',len(continous_feature))
```

```python
continous_feature
```

```python
## Lets analyze the continus feature by creating histograms to understand the distribution
for feature in continous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

## Exploratory Data Analysis -- Part2

```python
## We will be using logarithmic transformation

for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
for feature in continous_feature:
    data = dataset.copy()
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Outliers

```python
for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
## Find out the relationship between categorical variable and dependent feature SalesPrice
```

```python
from itertools import cycle, islice


for feature in categorical_features:
    data = dataset.copy()
#     data.groupby(feature)['SalePrice'].median().plot.bar()
    my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(data)))
    data.groupby(feature)['SalePrice'].median().plot(kind='bar',stacked=True, color=my_colors)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature+' vs SalePrice')
    plt.show()
```
